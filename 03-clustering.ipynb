```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
locations = pd.read_csv('locations.csv')
```

```python
locations = locations[(locations.longitude > -74.4) & (locations.longitude < -73.5) & (locations.latitude > 40.5) & (locations.latitude < 40.9)]
```

```python
locations.head()
```

```python
plt.figure(figsize=(32,20))
plt.scatter(locations.longitude, locations.latitude, alpha=0.2, s=.1)
```

```python
clusters = KMeans(n_clusters=25)
clusters.fit(locations.drop('pickup_datetime', axis=1))
```

```python
locations['neighborhood'] = clusters.labels_
```

```python
plt.figure(figsize=(32,20))
plt.scatter(locations.longitude, locations.latitude, alpha=0.2, s=.1, c=locations.neighborhood)
```
